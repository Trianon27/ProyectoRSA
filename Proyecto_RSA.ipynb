**Descripción Matemática**:

En primer lugar, a grandes rasgos para usar el RSA necesito un par de números que serán publicados al resto de personas involucradas. De tal forma que, si yo quiero que me envíen un mensaje de forma encriptada, podre descifrarlo con los números que entregué en un principio a los involucrados.

Dada esta situación tenemos un mensaje compuesto por n letras, a estas letras podemos hallarles una equivalencia numérica *m*, con este preámbulo, necesitamos dos números *e* y *N* que  surgen de un procedimiento explicado más adelante (donde su tamaño es considerablemente grande), de este modo tenemos un par de números de la forma (*e,N*), así pues para codificar el mensaje lo hacemos de la forma: 

$$m^{e} (mod N)$$

De tal modo que obtendremos un *r*, tal que:

$$r (mod m)$$

donde *r* seria mi mensaje cifrado con los valores conocidos, y el cual como se observa permite ocultar el verdadero mensaje que se tenía originalmente. Por otro lado si yo quiero descifrar el mensaje encriptado debo tomar el numero *N* y mi clave *d* que yo solo poseo y nadie más conoce de este modo con la pareja (*d,N*) realizo el mismo procedimiento  el cual fue expuesto en el inicio de esta sección, esta vez con la equivalencia  *r* que me han enviado, de tal manera que :

$$r^{d} (mod N) = m(mod N)$$

donde **m** representa el mensaje original encriptado.

Ahora bien cómo funciona y se halla esta clave única; como se había mencionado anteriormente se debe escoger dos primos suficientemente grandes *(p,q)* los cuales yo solo conozco y posteriormente se debe hallar el producto de ambos:

$$p*q=N$$

donde este número *N* producido por el producto de ambos, será el módulo en mi operación encriptadora y desencriptadora y eventualmente este número N debe ser publicado, posteriormente se empleará: 

$$\Phi(N)$$

Se establece una lista de números los cuales podemos encontrar factores comunes con el número N como otros no, en esta ocasión no se tendrá aquellos números los cuales comparten un factor común con N, de tal forma tenemos disponible los números **coprimos** con N, pero normalmente como se tiene un producto de dos primos muy grandes, hallar todos los coprimos para este número suele ser una tarea difícil, no obstante, esta misma cantidad de coprimos puede hallarse de la siguiente forma: 

$$\Phi(N)=(p-1)(q-1)$$

Procedemos a hallar **e**, el número de encriptación, el cual debe cumplir ciertas propiedades, en primer lugar* e* debe encontrarse entre: 

$$1<e<\Phi(N)$$

 y a su vez coprimo de:
 $$\Phi(N)$$
 
 De tal modo que este número *e* con el N hallado son los números que deben ser publicados
 
 Finalmente, así como halle un número encriptador debo hallar un número desencriptador *d* que se halla de tal forma que:
 $$d*e(mod\Phi(N))=1$$

 De tal forma se explica la idea base con la cual se encuentra conformado el algoritmo RSA

Recapitulando tenemos:

***1) Generación de par de claves***: Para generar un par de claves $$(KP; Kp)$$ En primer lugar se eligen aleatoriamente dos números primos grandes, p y q. después calculamos el producto $$n=p*q$$
Se escogerá ahora un número e comprimo con (p-1) y con (q-1). Este par de números (e,n) pueden ser conocidos por cualquiera, y constituyen la llamada ***clave pública***

 *e* por tanto debe tener un inverso módulo (p-1)(q-1) al que será llamado d. Se cumple que $$e*d=1 mod((p-1)(q-1))$$, que es lo mismo que decir que $$e*d=1+k (p-1)(q-1)$$ para algún entero *k*. La ***clave privada*** será el par *(d,n)*. Este número *d* debe mantenerse en secreto y solo será conocido por el propietario del par de claves.

 ***2) Cifrado del mensaje con clave pública***: Hay que hacer notar que con este algoritmo los mensajes que se cifran y descifran son números enteros de tamaño menor que n, no letras independientes como suele pasar con los cifrados César o Vigénere.

 Para obtener el mensaje cifrado *r* a partir del mensaje en claro *m*, se realiza la siguiente operación:
 $$r=m^{e} (mod n)$$

 ***3) Descifrado del mensaje con clave privada***: Para recuperar el mensaje original a partir del cifrado se realiza la siguiente operación:

 $$m=r^{d} (mod n)$$

 La justificación de lo anterior se encuentra dada por:

$$r^{d} (mod n)=(m^{e})^{d} (mod n)= (m^{1+k(p-1)(q-1)} (mod n) = (m^{(p-1)(q-1)})^{k}.m(mod n) [i]$$

Si se recuerda la función Euler 

 $$\Phi(n)=(p-1)(q-1)$$

y que en general, se tendrá que $$mdc(m,p)=mcd(m,q) = mcd(m,n)=1$$

Y por tanto según el teorema de Euler-Fermat, 

$$m^{\Phi(n)} \cong 1 (mod n) \rightarrow (m^{(p-1)(q-1)})^{k} \cong 1 (mod n) [ii] $$

De [i][ii] se obtiene que:
$$ r^{d} (mod n) =1.m(mod n) = m $$
para:
$$0<m<n$$

***Conmutatividad del cifrado y descifrado en RSA***:

Por las propiedades de la exponenciación modular, el cifrado y descifrado son conmutativos: 

$$m=(m^{e} mod n)^{d} mod n = m^d{e} mod n = (m^{d} mod n)^{e} mod n = m$$

Esto supone que, si cifrado *m* con la clave pública *e* y a continuación descifrando el resultado con la privada obtenemos de nuevo *m*, también podemos cifrar *m* con la clave privada d y descifrar el resultado con la clave pública *e*, volviendo a obtener *m*.

Esta propiedad es importante porque nos permite utilizar RSA no sólo para cifrar un mensaje, sino también para autenticar el mensaje.
